```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Covid Data.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['CLASIFFICATION_FINAL'].replace([2,3],1,inplace=True)
df['CLASIFFICATION_FINAL'].replace([4,5,6,7],0,inplace=True)
```

```python
df=df.drop(['DATE_DIED'],axis=1)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True)
```

```python
x=df[['USMER','PNEUMONIA','INTUBED','ICU','COPD','ASTHMA','INMSUPR','HIPERTENSION','PREGNANT','OTHER_DISEASE','OBESITY']]
y=df['CLASIFFICATION_FINAL']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
clf.fit(x,y)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
